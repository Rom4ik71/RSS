Далее были написаны основные функции программы:
Первая из них "parser" совершает процесс парсинга сайта и декодировки его данных. На входе она получает ссылку на RSS ленту, на выходе получаем данные с этой ленты.

```python
def parser(rssline):
    data=[]
    rss = rssline
    feed = feedparser.parse(rss)
    website = feed["feed"]["title"]
    for key in feed["entries"]:
        title = key["title"]
        link = unidecode.unidecode(key["link"])
        summary = key["summary"]
        data.append((title, link, website, summary))
    return data
```
